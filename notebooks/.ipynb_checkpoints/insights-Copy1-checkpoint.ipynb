```python
import snowflake.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
usuario = os.getenv('SNOWFLAKE_USER')
cuenta = os.getenv('SNOWFLAKE_ACCOUNT')
passwd = os.getenv('SNOWFLAKE_PIPELINESERVICE_PASSWORD')

conn = snowflake.connector.connect(
    user=usuario,
    password=passwd,
    account=cuenta,
    warehouse='COMPUTE_WH',
    database='INSTACART_DB',
    schema='CLEAN',
    protocol='https'
    )
#esto no es una funcion pq es mas rapido solo conectarse una vez y de ahi ejecutar todo en lugar de conectarse 
#multiples veces


def snowflake_to_pandas(table, dataframe,cur):

    sql = f"select * from \"{table}\";"
    cur.execute(sql)
    
    dfs = []
    
    for dfa in cur.fetch_pandas_batches():
        dfs.append(dfa)
    
    dataframe = pd.concat(dfs, ignore_index=True)
  
        
    return dataframe
```

```python
#importando las tablas y cerrando la conexion

df_fct_orders = snowflake_to_pandas("fct_orders", pd.DataFrame(), conn.cursor())
df_dim_products = snowflake_to_pandas("dim_products", pd.DataFrame(),conn.cursor())

conn.close()
```

```python
#funcion para obtener las metricas que yo quiero
def agg_std(df): 
    agg_functions = ['mean', 'std', 'min', 'max', 'var', 'median']
    
    agg_stats = {}
    
    for column in df.columns:
        column_stats = df[column].agg(agg_functions).to_dict()
        
        # Skewness and Kurtosis
        column_stats['skew'] = df[column].skew()
        column_stats['kurt'] = df[column].kurt()
        
        # Mode Handling
        mode_values = df[column].mode()
        column_stats['mode'] = mode_values.iloc[0] if not mode_values.empty else None
        
        # Quantiles
        quantiles = df[column].quantile([0.25, 0.5, 0.75]).to_dict()
        column_stats.update({f'quantile_{int(q*100)}': v for q, v in quantiles.items()})
        
        agg_stats[column] = column_stats
    
    agg_stats_df = pd.DataFrame(agg_stats)
    
    return agg_stats_df
```

```python
#Revisar si es que todos los tipos de los datos en cada columna son iguales
def check_type(df):
    for column in df.columns:
        same_type = df[column].apply(type).nunique() == 1
        print(f"All elements in the column '{column}' have the same data type: {same_type}")
```

```python
#Revisar si hay duplicados en cada columna del df
def check_duplicate(df):
    for column in df.columns:
        duplicate_count = df[column].duplicated().sum()
        print(f"There are {duplicate_count} duplicates in the {column}")
```

```python
#Crear un histogrma mas rapido
def hist(df,i,binz,title,xlabel,ylabel):
    plt.hist(df.iloc[:,i], bins=binz, edgecolor='black')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.show()
```

```python
#Funciones para crear gráficos más fácil
def box(df, column, title, ylabel):
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df.iloc[:, column], color="lightblue", width=0.5, fliersize=3)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()
```

```python
#Crear un violin plot mas rapido
def violin(df,i,title,ylabel):
    plt.figure(figsize=(6, 4))
    sns.violinplot(data=df.iloc[:,i], inner="box", color="lightblue")  # Ajustar suavidad
    plt.title(title)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.ylim(df.iloc[:,i].min(), df.iloc[:,i].max())
    plt.show()
```

```python
#una funcion donde se grafican dos lineas para comparar de una manera mas evidente los histogramas. Usa porcentajes
def dob(df, i, df2, i2, n, title, xlabel, ylabel,dia):
    freq1 = df.iloc[:, i].value_counts(normalize=True).sort_index() * 100
    freq2 = df2.iloc[:, i2].value_counts(normalize=True).sort_index() * 100
    
    plt.figure(figsize=(8, 5))

    plt.plot(freq1.index, freq1.values, marker='o', linestyle='-', color='#A19FA9', label="Todos los Días")

    plt.plot(freq2.index, freq2.values, marker='o', linestyle='-', color='lightblue', label=f"{dia}")

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    
    plt.grid(True, linestyle="--", alpha=0.6)
    
    plt.show()
```

```python
#funcion ipara comparar lineas exclusivamente para una donde se usa usuario
def dob_user(df, i, df2, i2, n, title, xlabel, ylabel):
    freq1 = df.iloc[:, i].value_counts().sort_index()
    freq2 = df2.iloc[:, i2].value_counts().sort_index()

    
    plt.figure(figsize=(8, 5))

    plt.plot(freq1.index, freq1.values, marker='o', linestyle='-', color='#A19FA9', label="Con order_number")

    plt.plot(freq2.index, freq2.values, marker='o', linestyle='-', color='lightblue', label="Sin order_number")

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    
    plt.grid(True, linestyle="--", alpha=0.6)
    
    plt.show()
```

```python
#funcion para crear un diccionario que contiene de manera ordenada un df y el dia de la semana, empezando por 
# domingo
def seg(df, i, r):
    dfs = {}
    for value in range(r):
        dfs[f"df_{df.columns[i]}_{value}"] = df[df.iloc[:, i] == value]
        
    return dfs
```

```python
#grafico de barras para usar con los dias de la semana ya que aqui colocamos el label dependiendo del valor
def bar_dow(df, i, title, xlabel, ylabel, coloreo=(None, None)):

    freq = df.iloc[:,i].value_counts()
    
    plt.bar(freq.index, freq.values, edgecolor='black')
    
    start, end = coloreo

    colors = ['#E4F5FD' if start <= idx <= end else '#A19FA9' for idx in range(len(list(freq.index)))]
    
    plt.bar(freq.index, freq.values, color=colors, edgecolor='black')
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.xticks(ticks=freq.index, labels=['Domingo', 'Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado'],rotation=45)
    
    plt.show()
```

```python
#funcion para obtener el color de la zona resaltada
def get_color(value,start,end):
    if start is not None and end is not None:
        return '#E4F5FD' if start <= value <= end else '#A19FA9'
    return '#A19FA9'
```

```python
#crear un grafico de barras que colorea la parte importante 
def bar(df, i, title, xlabel, ylabel, coloreo=(None, None)):
    freq = df.iloc[:, i].value_counts()
    
    start, end = coloreo

    colors = [get_color(val,start,end) for val in freq.index]
    
    plt.bar(freq.index, freq.values, color=colors, edgecolor='black')
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.show()
```

```python
#crea un grafico de barras colorinche usando porcentajes
def bar_perc(df, i, title, xlabel, ylabel, coloreo=(None, None)):
    if i < 0 or i >= df.shape[1]:  
        raise ValueError("Column index is out of range.")
    
    freq = df.iloc[:, i].value_counts(normalize=True) * 100
    indices = list(freq.index)
    
    start, end = coloreo

    colors = [get_color(idx,start,end) for idx in indices]

    plt.bar(indices, freq.values, color=colors, edgecolor='black')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.show()
```

## 1. Comportamiento de compra según día de la semana
○ Analizar la distribución de órdenes por cada día (0 = domingo, 1 = lunes, etc.).

Aquí vamos a usar el df_dim_products. Ahí tenemos el order_id y el order_dow. Hacemos que nos de por cada order único. El día se mantiene igual en la orden. Agrupamos porque hay ordenes que contienen más productos. No queremos que esas ordenes con mas productos sesguen nuestro análisis. Debemos de obtener métricas como skew y kurtosis. Tambien hacer un box plot.

```python
#agrupando por el fct order id para obtener el valor real de cada dia y que ordenes con muchos productos
#no inclinen el grafico
df_dim_grouped_orders = df_dim_products.groupby("fct_order_id").last().reset_index()

print(agg_std(df_dim_grouped_orders.iloc[:,5:6]))
```

El skew de order_dow nos indica que que es ligeramente asimétrio hacia la derecha mientras que la kurtosis negativa nos indica que su distribución es más plana que la normal. Podemos ver que el día más común es el domingo.

```python
#visualizando de una buena manera la frecuencia total de las compras
bar(df_dim_grouped_orders,5,"Distribución de la Frecuencia del Total de las Compras","Día de la Semana","Frecuencia de las Ventas",coloreo=(0,1))
```

Con el bar chart podemos entender mejor la distribución de la frecuencia de order_dow. Es más sencillo per cuales son los días más comunes y visualizar la forma de la gráfica. Domingo y lunes tienen más compras que los otros días.

```python
#visualizando la distribucion con un boxplot
box(df_dim_grouped_orders,5,"Distribución de las Ventas Durante Toda la Semana","Día de la Semana")
```

Viendo el box plot nos damos cuenta que no hay outliers. Esto es de esperarse ya que solamente hay 7 valores. Debería haber una distribución muy dispareja para que con 7 valores hayan outliers. Podemos visualizar a la mediana de mejor manera.

## 8. Proporción de pedidos que se vuelven a pedir para cada producto
○ Para cada producto, calcular cuántas veces es “reordenado” respecto al total de
pedidos del mismo.

Ordenamos por dim_product_id ahi y hacemos un count de reorder, en otro hacemos un count de reorder == 1 y dividimos

```python
#obteniendo el df del total de veces que aparece reorder, 0 y 1. Cambiando el nombre y ordenando
df_fct_grouped_products_total = df_fct_orders.groupby("dim_product_id").size().reset_index(name='reorder')
df_fct_grouped_products_total = df_fct_grouped_products_total.rename(columns = {"reorder":"reorder_total"})
df_fct_grouped_products_total = df_fct_grouped_products_total.sort_values(by="reorder_total",ascending=False)

#obteniendo el df de las veces que reorder = 1. Cambiando de nombre y ordenando
df_reorder_1 =df_fct_orders[df_fct_orders.iloc[:,3]==1]
df_fct_grouped_products_reordered_1 = df_reorder_1.groupby("dim_product_id").size().reset_index(name='reorder')
df_fct_grouped_products_reordered_1 = df_fct_grouped_products_reordered_1.sort_values(by="reorder",ascending=False)
df_fct_grouped_products_reordered_1 = df_fct_grouped_products_reordered_1.rename(columns = {"reorder":"reorder_1"})


df_overall = df_fct_grouped_products_total.merge(df_fct_grouped_products_reordered_1,on="dim_product_id",how="inner")
#esto join descarta aquellos que no tienen ningun reorder = 1.
df_overall["Proporción de Reorder"] = df_overall.iloc[:,-1]/df_overall.iloc[:,-2]
nombre_reorder = df_dim_products[df_dim_products["dim_product_id"].isin(df_overall["dim_product_id"])].groupby("dim_product_id").last().reset_index().iloc[:,[0,2,3,4]]

df_reorder_final = df_overall.merge(nombre_reorder, on="dim_product_id", how="inner")

#se realiza una division para obtener la proporcion y de ahi agrega una posicion
df_reorder_20 = df_reorder_final.sort_values(by="Proporción de Reorder",ascending=False).iloc[:20,:]

df_reorder_20["Posición"] = range(1, 21)

#Cambiando los nombres para que se vea mejor
df_reorder_20 = df_reorder_20.rename(columns = {"product_name":"Nombre del Producto"})
df_reorder_20 = df_reorder_20.rename(columns = {"department":"Departamento del Producto"})

print(df_reorder_20.iloc[:,[7,4,3,6]].to_string(index=False))
```

Podemos ver que los productos con mayor proporción de reorder tienen muy pocas ventas. Veamos cuales son los mejores que superen las 1000 ventas.

```python
#agrupando por la mayor cantidad de ventas y cambiando los nombres para que se vean mejor
reorder_top_20_final = df_reorder_final[df_reorder_final["reorder_total"]>1000].sort_values(by="Proporción de Reorder",ascending=False).iloc[:20,:]

reorder_top_20_final["Posición"] = range(1, 21)

reorder_top_20_final = reorder_top_20_final.rename(columns = {"product_name":"Nombre del Producto"})
reorder_top_20_final = reorder_top_20_final.rename(columns = {"department":"Departamento del Producto"})

print(reorder_top_20_final.iloc[:,[7,4,3,6]].to_string(index=False))
```

Estos son los productos con más de 1000 ventas que tienen la proporción de reorder más alta. Me parece interesante la presencia de "banana" en el top 5, considerando que también es el líder en ventas. Todos los productos pertenecen a departamentos relacionados a la comida. El único de estos productos que no pertenece a un departamento del top 20 de ventas es "Extra Fancy Unsalted Mixed Nuts" que pertenece al departamento de "snacks".

## 9. Proporción de productos pedidos que se vuelven a pedir para cada cliente
○ Cuántos productos vuelven a pedir los clientes en relación a la cantidad total de
productos comprados por cliente.

Agrupamos por user_id y hacemos count de reorder

```python
#Agrupando por usuario para obtener el total de reorder, cambiando el nombre y ordenando. Haciendo lo mismo 
#para cuando reorder = 1. Cambiando ciertos nombres para que se vea mejor

df_fct_user_reorder = df_fct_orders.groupby("user_id").size().reset_index(name='reorder')
df_fct_user_reorder= df_fct_user_reorder.rename(columns = {"reorder":"Total de Reorder"})
df_fct_user_reorder = df_fct_user_reorder.sort_values(by="Total de Reorder",ascending=False)

df_fct_reorder_1 = df_fct_orders[df_fct_orders.iloc[:,3]==1]
df_fct_user_reorder_1 = df_fct_reorder_1.groupby("user_id").size().reset_index(name='reorder')
df_fct_user_reorder_1 = df_fct_user_reorder_1.rename(columns = {"reorder":"reorder_1"})
df_fct_user_reorder_1 = df_fct_user_reorder_1.sort_values(by="reorder_1",ascending=False)



df_merge_user_reorder = df_fct_user_reorder.merge(df_fct_user_reorder_1,on="user_id",how="inner")

df_merge_user_reorder["Proporción de Reorder"] = df_merge_user_reorder.iloc[:,-1]/df_merge_user_reorder.iloc[:,-2]

user_top_20 = df_merge_user_reorder.sort_values(by="Proporción de Reorder",ascending=False).iloc[:20,:]

user_top_20["Posición"] = range(1, 21)

user_top_20 = user_top_20.rename(columns = {"user_id":"ID del Usuario"})

print(user_top_20.iloc[:,[4,0,1,3]].to_string(index=False))
```

Está claro que los usarios con mejor proporción apenas han realizado compras. Revisemos con un mayor volumen de compras. Con un reorder_total superior a 300

```python
#obteniendo los 20 usarios con mas ventas
user_top_20 = df_merge_user_reorder[df_merge_user_reorder["Total de Reorder"]>300].sort_values(by="Proporción de Reorder",ascending=False).iloc[:20,:]

user_top_20["Posición"] = range(1, 21)

print(user_top_20.iloc[:,[4,0,1,3]].to_string(index=False))
```

Es sumamente interesante ver la alta proporción de reorder que hay. Aún con 300 compras mínimas, el top 20 está compuesto por una proporción igual o mayor al 89%.
